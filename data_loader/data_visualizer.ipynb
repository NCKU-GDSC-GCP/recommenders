```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

## train_full.csv

```python
data_train_full = pd.read_csv("dataset/train_full.csv", low_memory=False)
data_train_full.head()
```

```python
data_train_full.info()
```

```python
msno.matrix(data_train_full)
```

```python
print(data_train_full["gender"].value_counts())
gender_null = np.count_nonzero(data_train_full['gender'].isnull())
print(gender_null)
gender_null / data_train_full.shape[0]
```

```python
sns.countplot("gender", data=data_train_full)
```

```python
print(data_train_full["location_type"].value_counts())
location_null = np.count_nonzero(data_train_full["location_type"].isnull())
print(location_null)

print("null ratio: ", location_null / data_train_full.shape[0])
```

```python
sns.countplot("location_type", data=data_train_full)
```

```python
sns.countplot(data_train_full["gender"], hue=data_train_full["location_type"])
```

```python
print(data_train_full["language"].value_counts())
language_null = np.count_nonzero(data_train_full["language"].isnull())
print(language_null)

language_null / data_train_full.shape[0] 
```

```python
openingTime_null = np.count_nonzero(data_train_full["OpeningTime"].isnull())
print(openingTime_null)

openingTime_null / data_train_full.shape[0]
```

```python
print(data_train_full["vendor_rating"].value_counts())
rating_null = np.count_nonzero(data_train_full["vendor_rating"].isnull())
print(rating_null)

rating_null / data_train_full.shape[0]
```

```python
sns.countplot("vendor_rating", data=data_train_full)
```
